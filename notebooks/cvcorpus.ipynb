# Audio File Analysis

```python
import librosa
import numpy as np

# Load an audio file
audio_file_path = "D:/cv-corpus-19.0-2024-09-13-pa-IN/cv-corpus-19.0-2024-09-13/pa-IN/clips/common_voice_pa-IN_21717904.mp3" 
y, sr = librosa.load(audio_file_path, sr=None)
```

```python
# Duration in seconds
duration = librosa.get_duration(y=y, sr=sr)
print("Duration (seconds):", duration)

# Mean and standard deviation of amplitude
mean_amplitude = np.mean(y)
std_amplitude = np.std(y)
print("Mean Amplitude:", mean_amplitude)
print("Amplitude Standard Deviation:", std_amplitude)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 4))
librosa.display.waveshow(y, sr=sr)
plt.title("Waveform")
plt.xlabel("Time (s)")
plt.ylabel("Amplitude")
plt.show()
```

# Spectrogram

```python
D = librosa.amplitude_to_db(np.abs(librosa.stft(y)), ref=np.max)
plt.figure(figsize=(10, 4))
librosa.display.specshow(D, sr=sr, x_axis='time', y_axis='log')
plt.colorbar(format="%+2.0f dB")
plt.title("Spectrogram")
plt.xlabel("Time (s)")
plt.ylabel("Frequency (Hz)")
plt.show()
```

# Mel Spectrogram

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
y, sr = librosa.load('D:/cv-corpus-19.0-2024-09-13-pa-IN/cv-corpus-19.0-2024-09-13/pa-IN/clips/common_voice_pa-IN_21717904.mp3', sr=22050)

# Generate the mel-spectrogram
mel_spec = librosa.feature.melspectrogram(y=y, sr=sr, n_mels=128)  # Corrected line
mel_spec_db = librosa.power_to_db(mel_spec, ref=np.max)

# Visualize the mel-spectrogram
plt.figure(figsize=(10, 4))
librosa.display.specshow(mel_spec_db, sr=sr, x_axis='time', y_axis='mel', fmax=8000)
plt.colorbar(format='%+2.0f dB')
plt.title('Mel-Spectrogram')
plt.show()
```

```python
import librosa
import librosa.display
import matplotlib.pyplot as plt
import numpy as np

# Load the audio file
audio_file_path = "D:/cv-corpus-19.0-2024-09-13-pa-IN/cv-corpus-19.0-2024-09-13/pa-IN/clips/common_voice_pa-IN_21717904.mp3"
y, sr = librosa.load(audio_file_path, sr=None)

# Extract MFCC features
mfccs = librosa.feature.mfcc(y=y, sr=sr, n_mfcc=13)  # Extract 13 MFCC coefficients

# Plot MFCC
plt.figure(figsize=(10, 4))
librosa.display.specshow(mfccs, x_axis='time', sr=sr)
plt.colorbar(format='%+2.0f dB')
plt.title('MFCC')
plt.xlabel('Time')
plt.ylabel('MFCC Coefficients')
plt.tight_layout()
plt.show()
```
